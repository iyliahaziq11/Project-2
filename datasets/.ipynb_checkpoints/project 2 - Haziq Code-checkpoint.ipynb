```python
# function to check for null values in dataframe
def check_for_null(df):
    for i in df.columns:
        if (len(df[df[i].isnull()==True]) > 0) or (len(df[df[i]==0]) > 0):
            print(f"{i}: {len(df[df[i].isnull()==True])/len(ames):.2f} NULL values, {len(df[df[i]==0])/len(ames):.2f} ZERO values")
```

# Types of feature selection:
1. Variance Threshold (VT)
    - from sklearn.feature_selection import VarianceThreshold
    - vt = VarianceThreshold()
2. Sequential Feature Selection (forward/backward selection)
3. Look for linked / connected features and drop one of them

1. drop very high % missing variable
2. amount of variation (i.e. too low; a feature must have high enough variation for the model to learn from it).
3. pairwise correlation; too correlated (>0.8) ==>drop 1 of the features    
4. feature correlation with target
5. forward/backward/stepwise selection
6. LASSO
7. Tree-based

# Ways to improve R^2 score:
1. add a interaction feature
2. remove predictor outliers

```python
null_list = []
null_val = []

for i in ames.columns:
    if len(ames[ames[i].isnull() == True]) >0:
        null_list.append(i)
        null_val.append(len(ames[ames[i].isnull() == True])/len(ames))
        
data = {"feature_name":null_list, "missing_counts":null_val}
missing_col = pd.DataFrame(data)
missing_col.sort_values(by="missing_counts", ascending=False, inplace=True)

# plotting bar graph to visualize column with highest num of mising values
plt.figure(figsize=(12,6))
plt.xticks(rotation = 90)
ax = sns.barplot(data=missing_col, x = "feature_name", y="missing_counts")
plt.xticks(rotation=50)
plt.xlabel('Feature names', labelpad=20, fontsize = 15)
plt.ylabel('% of values\n that are null', labelpad=50, fontsize = 15, rotation =0);
```

## Numerical Features

- Drop irrelevant columns
- Change date columns to reflect "Age" rather than the year, then convert to numerical

```python
# drop irrelevant columns
len(ames.select_dtypes(include='number').columns)
ames_num = (ames[ames.select_dtypes(include='number').columns]).drop(columns=['PID'])
# Id also has very high correlation with Yr Sold
ames_num.head()
# - Drop irrelevant columns
# - Change date columns to reflect "Age" rather than the year, then convert to numerical
import datetime
today = (datetime.date.today()).year
today

ames_num["age_built"] = today - ames['Year Built']
ames_num["age_remod/add"] = today - ames['Year Remod/Add']
ames_num["age_sold"] = today - ames["Yr Sold"]
ames_num["age_garage_built"] = today - ames["Garage Yr Blt"]

ames_num.drop(columns=['Year Built', 'Year Remod/Add', "Yr Sold"], inplace=True)

# ames_num["SF1 + SF2"] = ames_num["1st Flr SF"] + ames_num["2nd Flr SF"] # equivalent to GrLivArea - UnfSF
# ames_num['abvgrd bathroom'] =  ames_num['Full Bath'] + ames_num['Half Bath']
# ames_num['blwgrd bathroom'] = ames_num['Bsmt Full Bath']+ames_num['Bsmt Half Bath']


### Engineered columns to add (based on correlation to SalePrice)
ames_num["SF1 + SF2 + SFTotBsmt"] = ames_num["1st Flr SF"] + ames_num["2nd Flr SF"] + ames_num["Total Bsmt SF"] 
ames_num['deck + porch SF'] = ames_num['Wood Deck SF'] + ames_num['Open Porch SF']
ames_num['total bathroom'] = ames_num['Bsmt Full Bath']+ames_num['Bsmt Half Bath'] + ames_num['Full Bath'] + ames_num['Half Bath']


## Select top ten numerical features based on Corr value with SalePrice
## removed 'SF1 + SF2 + SFTotBsmt' and 'TotRms AbvGrd' due to high correlation with 'Gr Liv Area'
ames_num = ames_num.loc[:,
                        ['SalePrice',
                         'SF1 + SF2 + SFTotBsmt',
                         'Gr Liv Area',
                         'Garage Area',
                         'total bathroom',
#                          'TotRms AbvGrd',
                         'Mas Vnr Area',
                         "Fireplaces",
                         'deck + porch SF',
                         'BsmtFin SF 1',
                         'age_built',
                         'age_remod/add',
#                         'Lot Frontage',
#                         'Lot Area'
                        ]]
```
