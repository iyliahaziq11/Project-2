## <span style="color:orange">Custom Functions</span> 

```python
# function to check for null values in dataframe
def check_for_null(df):
    for i in df.columns:
        if (len(df[df[i].isnull()==True]) > 0) or (len(df[df[i]==0]) > 0):
            print(f"{i}: {len(df[df[i].isnull()==True])/len(ames):.2f} NULL values, \
                  {len(df[df[i]==0])/len(ames):.2f} ZERO values")

# function for multiple histogram plots
def subplot_hist(feature_list, main_dataframe):
    fig_row = math.ceil((len(feature_list)/3))
    fig_col = 3
    fig, axes = plt.subplots(fig_row, fig_col, figsize=(fig_col*5,fig_row*5))
    i=0
    for row in range(fig_row):
        for col in range(fig_col):
            axes[row,col].hist(ames[feature_list[i]], bins = 50)
            axes[row,col].set_title(feature_list[i])
            i+=1
            if i<len(feature_list):
                continue
            else:
                break

# function for multiple scatter plots
def subplot_scatter(X_feature_list, Y_feature, main_dataframe):
    fig_row = math.ceil(len(X_feature_list)/3)
    fig_col = 3
    fig, axes = plt.subplots(fig_row, fig_col, figsize=(fig_col*5,fig_row*5))
    i=0
    for row in range(fig_row):
        for col in range(fig_col):
            axes[row,col].scatter(ames[X_feature_list[i]], ames[Y_feature])
            axes[row,col].set_title(X_feature_list[i])
            i+=1
            if i<len(X_feature_list):
                continue
            else:
                break
                
# function for boxplot wrt SalePrice                
def subplot_boxplot(feature):
    ames1 = ames[[feature]]
    ames2 = ames[["SalePrice"]]
    df_mean= pd.concat((ames1, ames2),axis = 1)
    order_index = df_mean.groupby(feature).mean().sort_values(by='SalePrice').index
    plt.figure(figsize=(15,9))
    plt.xticks(rotation=60)
    sns.boxplot(data= df_mean, x=feature, y="SalePrice", order=order_index);
    print(ames[feature].value_counts())
```

- every neighborhood name will be changed to its respective ranking number
- the ones with higher neighborhood_ranking number will be assigned a higher ordinal number ranging from 1 to 4.

```python
# !!! CHANGED ORIGINAL !!! 


ames['neighborhood_ranking'] = ((ames["Overall Qual"]/ames["Overall Qual"].mean()) +
                                (ames["Overall Cond"]/ames["Overall Cond"].mean()) +
                                (ames["Exter Qual"]/ames["Exter Qual"].mean()) +
                                (ames["Exter Cond"]/ames["Exter Cond"].mean()) +
                                (ames["Bsmt Qual"]/ames["Bsmt Qual"].mean()) +
                                (ames["Bsmt Cond"]/ames["Bsmt Cond"].mean()) +
                                (ames["Heating QC"]/ames["Heating QC"].mean()) +
                                (ames["Kitchen Qual"]/ames["Kitchen Qual"].mean()) +
                                (ames["Garage Qual"]/ames["Garage Qual"].mean()) +
                                (ames["Garage Cond"]/ames["Garage Cond"].mean()) 
#                                 (ames["Pool QC"]/ames["Pool QC"].mean())
                               )
```

```python
# to show that ranking follows somewhat of a normal distribution
plt.figure(figsize=(10,6))

sns.distplot(x=ames['neighborhood_ranking'], bins=40);
```

```python
# df_nbr = df_nbr.reset_index().drop(columns='SalePrice')
df_nbr = ames.groupby('Neighborhood').mean()[['neighborhood_ranking']].sort_values(
    'neighborhood_ranking').reset_index()
```

df_nbr.to_csv('neighborhood ranking')

```python
# !!! CHANGED ORIGINAL !!! 

category_4th = list(df_nbr.loc[0:6,"Neighborhood"])
category_3rd = list(df_nbr.loc[7:13,"Neighborhood"])
category_2nd = list(df_nbr.loc[14:20,"Neighborhood"])
category_1st = list(df_nbr.loc[21:28,"Neighborhood"])

for i in range(len(ames)):
    if ames.loc[i,'Neighborhood'] in category_4th:
        ames.loc[i,'neighborhood_ranking'] = 1
    elif ames.loc[i,'Neighborhood'] in category_3rd:
        ames.loc[i,'neighborhood_ranking'] = 2
    elif ames.loc[i,'Neighborhood'] in category_2nd:
        ames.loc[i,'neighborhood_ranking'] = 3
    elif ames.loc[i,'Neighborhood'] in category_1st:
        ames.loc[i,'neighborhood_ranking'] = 4
        
ames['neighborhood_ranking'] = ames['neighborhood_ranking'].astype('int')
feature_num = feature_num + ['neighborhood_ranking']

if "Neighborhood_ranking" in feature_num:
    feature_num.remove("Neighborhood_ranking")
    
if "Neighborhood" in feature_cat:
    feature_cat.remove("Neighborhood")
    
ames.head()
ames.groupby('Neighborhood').mean()[['neighborhood_ranking']].sort_values(by='neighborhood_ranking')
```
