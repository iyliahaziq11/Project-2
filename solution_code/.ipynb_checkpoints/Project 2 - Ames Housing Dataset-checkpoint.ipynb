# Project 2: Ames Housing Data and Kaggle Challenge

# Problem Statement

- Creating a regression model based on the Ames Housing Dataset. This model will predict the price of a house at sale.
- The main target audience will be property agents having trouble evaluating the price of each property in Ames.

# Background

Ames is a city in the state of Iowa of Unite States. In 2020, Ames had a population of about 66,000. Majority of the population are in the 15-30 years age range (roughly 50%). As of 2010, which is in the same period as our project, there were 23,876 housing units. Our model will be based on about 2000 datasets , which is roughly 10% of the total housing units. The average household size was 2.25 and the average family size was 2.82. In 2010, Ames had a real per capita income of $30,128. ([*source*](https://www.deptofnumbers.com/income/iowa/ames/)). These numbers give an indication on the types of house built in Ames and how it is priced, which we will look at later. ([*source*](https://en.wikipedia.org/wiki/Ames,_Iowa#cite_note-wwwcensusgov-15))

Another interesting point to note is that the number of housing units outpaced the growth in total households. This could mean that supply is more than demand. However one cannot be sure, especially with data showing that 30% of householders are living alone ([*source*](https://www.icip.iastate.edu/sites/default/files/2010census/2010census_1901855.pdf)). So the increased supply of housing units could be due to individuals who chose to stay alone. Either way, we will attempt to predict the housing price based on multiple features.

Equipped with this knowledge, we aim to build a good model prediction for house sales price based on datasets of various house features from 2006-2010.

```python
# function to check for null values in dataframe
def check_for_null(df):
    for i in df.columns:
        if (len(df[df[i].isnull()==True]) > 0) or (len(df[df[i]==0]) > 0):
            print(f"{i}: {len(df[df[i].isnull()==True])/len(ames):.2f} NULL values, \
                  {len(df[df[i]==0])/len(ames):.2f} ZERO values")

# function for multiple histogram plots
def subplot_hist(feature_list, main_dataframe):
    fig_row = math.ceil((len(feature_list)/3))
    fig_col = 3
    fig, axes = plt.subplots(fig_row, fig_col, figsize=(fig_col*5,fig_row*5))
    i=0
    for row in range(fig_row):
        for col in range(fig_col):
            axes[row,col].hist(ames[feature_list[i]], bins = 50)
            axes[row,col].set_title(feature_list[i])
            i+=1
            if i<len(feature_list):
                continue
            else:
                break

# function for multiple scatter plots
def subplot_scatter(X_feature_list, Y_feature, main_dataframe):
    fig_row = math.ceil(len(X_feature_list)/3)
    fig_col = 3
    fig, axes = plt.subplots(fig_row, fig_col, figsize=(fig_col*5,fig_row*5))
    i=0
    for row in range(fig_row):
        for col in range(fig_col):
            axes[row,col].scatter(ames[X_feature_list[i]], ames[Y_feature])
            axes[row,col].set_title(X_feature_list[i])
            i+=1
            if i<len(X_feature_list):
                continue
            else:
                break
                
# function for boxplot wrt SalePrice                
def subplot_boxplot(feature):
    ames1 = ames[[feature]]
    ames2 = ames[["SalePrice"]]
    df_mean= pd.concat((ames1, ames2),axis = 1)
    order_index = df_mean.groupby(feature).mean().sort_values(by='SalePrice').index
    plt.figure(figsize=(15,9))
    plt.xticks(rotation=60)
    sns.boxplot(data= df_mean, x=feature, y="SalePrice", order=order_index);
    print(ames[feature].value_counts())
```

### Data Cleaning and EDA - <span style="color:red">imputing null values</span> 
- Initial observations show that there were many null values and possibly even incorrect data. 26 features had at least 1 null values.
- Upon further investigation, it was observed that 15 were categorical and 11 were numerical.
- Categorical Features: Null values were found to be due to the lack of this feature. For example, null values for "Garage Quality" were due to the fact that the house had no garage. Hence these null values were imputed with "None" to reflect the absence of the feature.
- Numerical Features: Similarly, most null values were found to be due to the lack of this feature. As such, these null values were imputed with 0. However, some null values could not be explained. For these null values, they were imputed with its respective median value.

```python
import math

num_features_to_drop =[
    'Garage Area',
    '1st Flr SF',
    '2nd Flr SF',
    'Total Bsmt SF',
    'Mas Vnr Area',
#     'Overall Qual',
    'Kitchen Qual',
    'Bsmt Qual',
    'Exter Qual',
    'Garage Cond',
    'age_Garage Yr Blt',
    'age_Year Remod/Add',
    'Garage Yr Blt',
    'Year Remod/Add',
    'Year Built',
    'age_Yr Sold',
    'BsmtFin SF 2',
    'Bsmt Half Bath',
#     'Bsmt Unf SF'
]

feature_num = [feature for feature in ames.columns 
               if ames[feature].dtypes != 'object' and feature not in num_features_to_drop]
```

### <span style="color:red">Apply log to skewed numerical features </span> 

- log cannot be applied to 0 values, hence we filter features without 0 values and then apply log
- comparison before and after log transformation

```python
#from the above charts, we observe that the 3 skewed features that can be log-ed are Lot Frontage, Lot Area, Gr Liv Area

fig,axes = plt.subplots(2,3,figsize=(30,15))

axes[0,0].hist(ames['Lot Frontage'], bins=50)
axes[0,0].set_title("Lot Frontage", fontsize=20)
axes[0,1].hist(ames['Lot Area'], bins=50)
axes[0,1].set_title("Lot Area", fontsize=20)
axes[0,2].hist(ames['Gr Liv Area'], bins=50)
axes[0,2].set_title("Gr Liv Area", fontsize=20)

axes[1,0].hist(np.log(ames['Lot Frontage']), bins=50)
axes[1,0].set_title("Lot Frontage", fontsize=20)
axes[1,1].hist(np.log(ames['Lot Area']), bins=50)
axes[1,1].set_title("Lot Area", fontsize=20)
axes[1,2].hist(np.log(ames['Gr Liv Area']), bins=50)
axes[1,2].set_title("Gr Liv Area", fontsize=20);
```

##### We identify 3 features with skewed values and would benefit from logarithmic transformation. The above shows the charts of the features pre-log and post-log. We add the logarithmic-transformed features, and remove the original features
- Lot Frontage -> log_Lot Frontage
- Lot Area -> log_Lot Area
- Gr Liv Area -> log_Gr Liv Area

```python
# apply log to only the selected 3 features, since there are improvement to skewness
for feature in ['Lot Frontage', 'Lot Area', 'Gr Liv Area']:
    ames['log_' + feature] = np.log(ames[feature])
    feature_num.append('log_' + feature)
    if feature in feature_num:
        feature_num.remove(feature)
```

### Preprocessing and Modelling - <span style="color:firebrick; font-size: 15px">Categorical Features</span> 
- Remove features with 1 value that is highly dominating. After iterating the model using 80%, 85% and 90% , we found that by removing features that have 1 value taking up 85% of the proportions, the model gave the best score.
- We observed there are too many neighborhood classes that may lead to noises and therefore overfitting of model and hence we try to feature engineer it into an ordinal form.
  - Naturally, there would be neighbourhoods that are richer than others and hence there is an order wrt SalePrice as shown in the swarmplot below.
  - We try to infer this ordinal feature from other quality/condition features.
- We engineer a new feature called 'neighborhood_ranking' which is based on the sum of averages of all quality and condition features (except Pool QC due to its large missing data) for each neighborhood. We then split these ranked neighborhoods into 4 categories with ordinal values ranging from 1 to 4.


```python
feature_cat = [feature for feature in ames.columns if ames[feature].dtypes == "object"]
# subplot_hist(feature_cat, ames)
# remove unique number/id features
feature_cat.remove('Id')
feature_cat.remove('PID')

def subplot_hist2(feature_list, main_dataframe):
    import math
    fig_col = 5
    fig_row = math.ceil(len(feature_list)/fig_col)
    fig, axes = plt.subplots(fig_row, fig_col, figsize=(fig_col*5,fig_row*5))
    i=0
    for row in range(fig_row):
        for col in range(fig_col):
            axes[row,col].hist(ames[feature_list[i]], bins = 50)
            axes[row,col].set_title(feature_list[i])

            i+=1
            if i<len(feature_list):
                continue
            else:
                break

subplot_hist2(feature_cat, ames)
```

##### from above charts, we identify features that have > 85% 1 dominant value
- 80%, 85%, 90% were tested and 85% gave the best modelled R score


```python
data =[]
for i in range(3,len(feature_cat)):
    x=ames[feature_cat[i]].copy().value_counts(ascending=False).reset_index()
    pct=np.round(x.iloc[0,1]/x.iloc[:,1].sum(),2)
#     print(feature_cat[i], x.iloc[0,0], x.iloc[0,1], pct)
    if pct >0.85:
        data.append(feature_cat[i])
print(f"These are the columns that will be dropped: {data}")
# remove these columns 
feature_cat = [feature for feature in feature_cat if feature not in data]

# additionally after running multiple model iterations, we found that dropping these 4 features increased the score slightly 
feature_cat.remove('Street')
feature_cat.remove('BsmtFin Type 1')
feature_cat.remove('BsmtFin Type 2')
feature_cat.remove('Exterior 2nd')
```

###### <span style="color:firebrick; font-size: 15px">Transform Neighborhood feature into ordinal feature</span> 


- We observed there are too many neighborhood classes that may lead to noises and therefore overfitting of model and hence we try to feature engineer it into an ordinal form.
- Logically, there would be neighbourhood that are richer than others and hence there is an order wrt SalePrice
- We try to infer this ordinal feature from other quality/condition features

```python
# boxplot and swarmplot chart to show the variations in SalePrice between neighborhoods 
# which justifies our reasons above
ames1 = ames[["Neighborhood"]]
ames2 = ames[["SalePrice"]]
df_neighborhood_mean= pd.concat((ames1, ames2),axis = 1)
order_index = df_neighborhood_mean.groupby('Neighborhood').mean().sort_values(by='SalePrice').index

plt.figure(figsize=(12,6))
plt.xticks(rotation=60)
sns.boxplot(data= df_neighborhood_mean, x='Neighborhood', y="SalePrice", order=order_index)
plt.xlabel("Neighborhood", labelpad=20, fontsize=15)
plt.ylabel("SalePrice", labelpad=20, fontsize=15)
plt.title("Boxplot of saleprice in each neighborhood", fontsize=15);


plt.figure(figsize=(12,6))
plt.xticks(rotation=60)
sns.stripplot(data= df_neighborhood_mean, x='Neighborhood', y="SalePrice", order=order_index)
plt.xlabel("Neighborhood", labelpad=20, fontsize=15)
plt.ylabel("SalePrice", labelpad=20, fontsize=15)
plt.title("Swarmplot of mean saleprice in each neighborhood", fontsize=15);
```

##### Creating a neighborhood_ranking feature based on quality/condition features
- every neighborhood name will be changed to its respective ranking number
- the ones with higher neighborhood_ranking number will be assigned a higher ordinal number ranging from 1 to 4.

```python
ames['neighborhood_ranking'] = ((ames["Overall Qual"]/ames["Overall Qual"].mean()) +
                                (ames["Overall Cond"]/ames["Overall Cond"].mean()) +
                                (ames["Exter Qual"]/ames["Exter Qual"].mean()) +
                                (ames["Exter Cond"]/ames["Exter Cond"].mean()) +
                                (ames["Bsmt Qual"]/ames["Bsmt Qual"].mean()) +
                                (ames["Bsmt Cond"]/ames["Bsmt Cond"].mean()) +
                                (ames["Heating QC"]/ames["Heating QC"].mean()) +
                                (ames["Kitchen Qual"]/ames["Kitchen Qual"].mean()) +
                                (ames["Garage Qual"]/ames["Garage Qual"].mean()) +
                                (ames["Garage Cond"]/ames["Garage Cond"].mean()) 
#                                 (ames["Pool QC"]/ames["Pool QC"].mean())
                               )
```

```python
# neighborhood_ranking distribution to show that ranking follows somewhat of a normal distribution
plt.figure(figsize=(10,6))

sns.distplot(x=ames['neighborhood_ranking'], bins=40)
plt.xlabel("neighborhood ranking", labelpad=20, fontsize=15)
plt.ylabel("density", labelpad=20, fontsize=15);
plt.title('Distribution of neighborhood ranking');
```

```python
df_nbr = ames.groupby('Neighborhood').mean()[['neighborhood_ranking','SalePrice']].sort_values(
    'neighborhood_ranking').reset_index()
```

```python
category_4th = list(df_nbr.loc[0:6,"Neighborhood"])
category_3rd = list(df_nbr.loc[7:13,"Neighborhood"])
category_2nd = list(df_nbr.loc[14:20,"Neighborhood"])
category_1st = list(df_nbr.loc[21:28,"Neighborhood"])

for i in range(len(ames)):
    if ames.loc[i,'Neighborhood'] in category_4th:
        ames.loc[i,'neighborhood_ranking'] = 1
    elif ames.loc[i,'Neighborhood'] in category_3rd:
        ames.loc[i,'neighborhood_ranking'] = 2
    elif ames.loc[i,'Neighborhood'] in category_2nd:
        ames.loc[i,'neighborhood_ranking'] = 3
    elif ames.loc[i,'Neighborhood'] in category_1st:
        ames.loc[i,'neighborhood_ranking'] = 4
        
ames['neighborhood_ranking'] = ames['neighborhood_ranking'].astype('int')
feature_num = feature_num + ['neighborhood_ranking']

if "Neighborhood_ranking" in feature_num:
    feature_num.remove("Neighborhood_ranking")
    
if "Neighborhood" in feature_cat:
    feature_cat.remove("Neighborhood")
```
